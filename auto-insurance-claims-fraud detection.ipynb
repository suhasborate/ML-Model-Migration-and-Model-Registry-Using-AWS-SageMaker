```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the dataset into jupyter notebook:
df=pd.read_csv('C:\\Users\\suhas.borate\\Desktop\\insurance_claims.csv')

#Setting to display all columns
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
# Few columns contains '?' in dataset, so Replacing '?' by 'NaN' values: 'NaN' stands for 'Not a Number'(i.e. missing value)
df = df.replace('?',np.NaN)
```

```python
# Checking for missing values in dataset:
df.isna().sum()
```

```python
# Missing value imputation:
df['collision_type'].fillna(df['collision_type'].mode()[0],inplace=True)
df['property_damage'].fillna(df['property_damage'].mode()[0],inplace=True)
df['police_report_available'].fillna(df['police_report_available'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
df.drop(['policy_number','policy_bind_date','insured_zip','incident_date'],axis=1,inplace=True)
```

```python
df.shape
```

```python
# Checking for datatype of each column:
df.dtypes
```

```python
s = (df.dtypes == 'object')
cate_cols = list(s[s].index)

print("Categorical Features:")
print(cate_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cate_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
# Checking for 'fraud_reported':
df['fraud_reported'].value_counts()
```

```python
x = df.drop('fraud_reported', axis = 1)
y = df['fraud_reported']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
print("x_train shape:",x_train.shape)
print("x_test shape:",x_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```

```python
import xgboost as xgb
print(xgb.__version__)
```

```python
bt = xgb.XGBClassifier(max_depth=5, learning_rate=0.2, n_estimators=10, objective='binary:logistic')  # Setup xgboost model

bt.fit(x_train, y_train, verbose=False)  # Train it to our data
```

```python
# Predict and compare with real labels
y_pred = bt.predict(x_test)
y_pred
```

```python
y_test.to_numpy()
```

```python
# Creating new dataset, which contains Actual values and Predicted values only:
df1 = pd.DataFrame({'Actual_Fraud ':y_test.to_numpy(),'Predicted_Fraud':y_pred})
pd.set_option('display.max_rows', None)
df1
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print (accuracy_score(y_test,y_pred))
```

```python
print (confusion_matrix(y_test, y_pred))
```

```python
plt.clf()
plt.imshow(confusion_matrix(y_test, y_pred), interpolation='nearest', cmap=plt.cm.Wistia)
classNames = ['Positive(0)','Negative(1)']
plt.title('Fraud or Non-Fraud Confusion Matrix - Test Data')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
s = [['TP','FN'], ['FP', 'TN']]
for i in range(2):
    for j in range(2):
        plt.text(j,i, str(s[i][j])+" = "+str(confusion_matrix(y_test,y_pred)[i][j]))
plt.show()
```

```python
# Let's Print Classification report:
print (classification_report(y_test,y_pred))
```
